```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# prompt: load dataset with pandas file in xslx

df = pd.read_excel('/content/drive/MyDrive/dataset1/Crime_data.xlsx')
```

```python
df
```

```python
# with sns plot a heatmap for see null values

plt.figure(figsize=(10, 8))
sns.heatmap(df.isnull(), cmap='viridis')
plt.title("Heatmap for Null Values")
plt.show()
```

```python
# with pandas lets drop all null values of df

df.dropna(inplace=True)
```

```python
# now lets search duplicates in df

duplicates = df[df.duplicated()]
print(duplicates)
```

```python
# prompt: dowload df with pandas

df.to_csv('/content/drive/MyDrive/dataset1/Crime_data_cleaned.csv')
```
